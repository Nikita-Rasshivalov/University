```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import copy as cop

from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,precision_score, recall_score,f1_score,plot_roc_curve
from sklearn.metrics import log_loss
```

```python
df = pd.read_csv('flavors_of_cacao.csv')
df.head()
```

```python
df=df.rename(columns={'Company':'company','Specific Bean Origin or Bar Name':'specificBean','REF':'ref','Review\nDate':'reviewDate','Cocoa\nPercent':'cocoaPercent','Company\nLocation':'companyLocation','Rating':'rating','Bean\nType':'beanType','Broad Bean\nOrigin':'broadBeanOrigin'})
print(df.columns)

```

```python
print("Bean Type:")
print(df['beanType'].isnull().sum())
print(round(df['beanType'].isnull().sum() * 100 / len(df),1),"%")
print("Broad Bean Origin:")
print(df['broadBeanOrigin'].isnull().sum())
print(round(df['broadBeanOrigin'].isnull().sum() * 100 / len(df),1),"%")
```

```python
df['cocoaPercent'] = df['cocoaPercent'].str.replace('%','').astype(float)/100
sns.displot(data=df,x='cocoaPercent',kde=True,color='green')
plt.show()
fig, ax = plt.subplots(figsize=[16,4])
sns.distplot(df['cocoaPercent'], ax=ax)
plt.show
```

```python
print("Amount of unique values:")
print('company')
print(len(df['company'].unique()))
print('Company Location')
print(len(df['companyLocation'].unique()))
print('Rating')
print(len(df['rating'].unique()))
print('Specific Bean Origin or Bar Name')
print(len(df['specificBean'].unique()))
print('REF')
print(len(df['ref'].unique()))
print('Cocoa Percent')
print(len(df['cocoaPercent'].unique()))
print("Amount:"+str(len(df)))

```

```python
k = 0
j = 0
for  i in range(len(df)):
    if df.loc[i,'cocoaPercent'] >= 0.70:
        df.loc[i,'cocoaPercent'] = 1
        k+=1  
    else:
        df.loc[i,'cocoaPercent'] = 0
        j+=1
print(k,j)
```

```python
def code_mean(data,category,real):
    return (data[category].map(data.groupby(category)[real].mean()))
df3 = cop.deepcopy(df)
df3['location_rating'] = code_mean(df3,'companyLocation','rating')
df3['broadBeanOrigin_rating'] = code_mean(df3,'broadBeanOrigin','rating')
df3['beanType_rating'] = code_mean(df3,'beanType','rating')
df3 =df3.drop(['company','specificBean','broadBeanOrigin','ref','companyLocation','beanType'],axis=1)
df3 = df3.dropna(how='any',axis=0) 
print(df3.corr())
```

```python
from sklearn.model_selection import train_test_split
dataX = df3.drop(['cocoaPercent'],axis=1) 
dataY = df3['cocoaPercent'] #отбор классифицируемого признака\
train_x, test_x, train_y, test_y = train_test_split(dataX,dataY,test_size = 0.3,random_state = 11) #разделение выборки на обучающую и тестовую
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(metric='euclidean') #создание классификатора k-ближайших соседей с использованием манхеттенской метрики
knn.fit(train_x, train_y) #обучение классификатора
#print(knn.score(test_x,test_y)) 
scores = cross_val_score(knn, dataX, dataY, cv=4)
print("Cross-validation result:"+str(round(scores.mean(),3)))
plt.figure(figsize=(10,10))
sns.heatmap(df3.corr(),annot=True,fmt='.1f')
plt.show()
```

```python
pred_y = knn.predict(test_x)
mat = confusion_matrix(test_y[0:1000], pred_y[0:1000])
print(mat)

print("Accuracy:"+str(round(accuracy_score(pred_y,test_y),3)))
print("Precision:"+str(round(precision_score(pred_y, test_y,average='micro'),3)))
print("Recall:"+str(round(recall_score(pred_y, test_y,average='micro'),3)))
#x = list(range(0, len(test_y)))
#fig, ax = plt.subplots()
#plt.plot(x[0:1000],test_y[0:1000],color="red")
#plt.plot(x[0:1000],pred_y[0:1000],color="blue")
#ax.legend(labels=["Initial","Predict"])
#plt.show()
```

```python
predicts = knn.predict_proba(test_x)    
print(log_loss(test_y,predicts))
```
