```python
dataX=df.drop(['thalach','fbs'],axis=1)
dataY=df['thalach']
for  i in range(len(dataY)):
    if df.loc[i,'thalach'] >= 160:
        df.loc[i,'thalach'] = 1
    else:
        df.loc[i,'thalach'] = 0 
dataX = (dataX - np.min(dataX)) / (np.max(dataX) - np.min(dataX)).values
x_train,x_test,y_train,y_test=train_test_split(dataX,dataY,test_size=0.2,random_state=0)
```

```python
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)
score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
selectedIndex=1
print(df.iloc[selectedIndex])
print('Real target='+str(df['thalach'].iloc[selectedIndex]))
predict=logisticRegr.predict([dataX.iloc[selectedIndex]])
predict2 = logisticRegr.predict_proba([dataX.iloc[selectedIndex]])
print("Predict="+str(predict[0]))
print("The probability of having the disease="+str(round(predict2[0][1],3)*100)+"%")
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression(fit_intercept=True , normalize=True)
reg.fit(x_train , y_train)
y_pred = reg.predict(x_test)
print(reg.score(x_train,y_train))

```
